```python
import os
import cv2
import pickle
from PIL import Image as im
from PIL import ImageTk
from IPython.display import Image
```

# Importing the image

```python
image = cv2.imread("dog.jpg")
Image('C:\\Users\\nanth\\dog.jpg',width=250,
    height=50)
```

```python
#image width and height
print("Width of image:",image.shape[1])
print("Height of image:",image.shape[0])
```

# Adding grey filter

```python
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
data = im.fromarray(grey_image).resize(size=(500,400))
data
```

# Adding Inverted filter

```python
inverted_image = 250 - grey_image
data = im.fromarray(inverted_image).resize(size=(500,400))
data
```

# Adding Blur effect

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
inverted_blurred_image = 255 - blurred
data = im.fromarray(inverted_blurred_image).resize(size=(500,400))
data
```

# Final Pencil Sketch

```python
pencil_sketch = cv2.divide(grey_image, inverted_blurred_image, scale=256.0)
data = im.fromarray(pencil_sketch).resize(size=(500,400))
data
```

# Exporting the Image

```python
# Saving the image
cv2.imwrite('Final_pencil_sketch.png',pencil_sketch)
```
